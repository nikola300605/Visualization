```python
import pandas as pd

def describe_attribute_distributions(df):
    """
    Prints the distribution and frequency counts of each attribute
    in a given DataFrame.

    Parameters:
    df (pd.DataFrame): The dataset containing attributes to analyze.
    """

    for col in df.columns:
        print("="*80)
        print(f"ATTRIBUTE: {col}")
        print("-"*80)

        # Numerical columns → summary statistics + histogram-like data
        if pd.api.types.is_numeric_dtype(df[col]):
            print("Type: Numeric")
            print("\nSummary statistics:")
            print(df[col].describe())

            print("\nValue frequency (binned):")
            print(df[col].value_counts(bins=10, dropna=False))  # histogram-like bins

        # Categorical columns → frequency table
        else:
            print("Type: Categorical")
            print("\nValue frequency:")
            print(df[col].value_counts(dropna=False))

        print("\n")
```

```python
import pandas as pd

def missing_summary(df):
    """
    Returns:
    - missing_per_attribute: number of missing values for each column
    - missing_per_row: number of missing values in each entry (row)
    """
    return {
        "missing_per_attribute": df.isna().sum(),
        "missing_per_row": df.isna().sum(axis=1)
    }
```

```python
import pandas as pd
from pathlib import Path

data_folder = Path("C:/Users/krist/OneDrive/Visualisation project")

files = [
    "communications_data.csv",
    "demographics_data.csv",
    "economy_data.csv",
    "energy_data.csv",
    "geography_data.csv",
    "government_and_civics_data.csv",
    "transportation_data.csv",
]

dfs = []

for fname in files:
    path = data_folder / fname
    df = pd.read_csv(path)
    
    if "Country" in df.columns:
        df = df.set_index("Country")
    else:
        raise ValueError(f"'Country' column not found in {fname}")
    
    prefix = fname.replace("_data.csv", "")
    df = df.add_prefix(prefix + "_")
    
    dfs.append(df)

combined = pd.concat(dfs, axis=1, join="outer")

missing_per_country = combined.isna().sum(axis=1)

missing_per_country_sorted = missing_per_country.sort_values(ascending=False)

print("Top countries with the most missing values:")
print(missing_per_country_sorted.head(20))
```
